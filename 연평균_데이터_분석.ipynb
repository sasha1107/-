```python
import pandas as pd
import numpy as np
from functools import reduce

file_1 = r"C:\Users\sasha\Documents\Bigdata\2001\2001년01분기.csv"
file_2 = r"C:\Users\sasha\Documents\Bigdata\2001\2001년02분기.csv"
file_3 = r"C:\Users\sasha\Documents\Bigdata\2001\2001년03분기.csv"
file_4 = r"C:\Users\sasha\Documents\Bigdata\2001\2001년04분기.csv"

for i in range(1,5):
    globals()['df_{}'.format(i)] = pd.read_csv( globals()['file_{}'.format(i)], encoding='utf-8')
```

###  지역별 연평균 구하기

```python
for i in range(1,5):
    globals()['mtp_{}'.format(i)] = pd.DataFrame(globals()['df_{}'.format(i)].groupby('시도명')['monthly_PM10'].mean())
    globals()['do_{}'.format(i)] = pd.DataFrame(globals()['df_{}'.format(i)].groupby(['시도명','시군구명'])['monthly_PM10'].mean())
    
    globals()['mtp_{}'.format(i)] = globals()['mtp_{}'.format(i)].reset_index() #index --> column으로
    globals()['do_{}'.format(i)] = globals()['do_{}'.format(i)].reset_index() #index --> column으로
    globals()['mtp_{}'.format(i)] = globals()['mtp_{}'.format(i)].loc[(globals()['mtp_{}'.format(i)]['시도명'] =='서울')|(globals()['mtp_{}'.format(i)]['시도명'] =='부산')|(globals()['mtp_{}'.format(i)]['시도명'] =='대구')|(globals()['mtp_{}'.format(i)]['시도명'] =='인천')|(globals()['mtp_{}'.format(i)]['시도명'] =='대전')|(globals()['mtp_{}'.format(i)]['시도명'] =='광주')|(globals()['mtp_{}'.format(i)]['시도명'] =='울산')|(globals()['mtp_{}'.format(i)]['시도명'] =='세종')]
```

```python
do_dfs = [do_1,do_2, do_3, do_4]
mtp_dfs = [mtp_1, mtp_2, mtp_3, mtp_4]
do_df = reduce(lambda left, right: pd.merge(left, right, on=['시도명','시군구명']), do_dfs)
mtp_df = reduce(lambda left, right: pd.merge(left, right, on='시도명'), mtp_dfs)
```

```python
do_df.loc[:,'annual_PM10'] = do_df.mean(axis=1) #분기별 평균 --> 연별 평균으로
mtp_df.loc[:,'annual_PM10'] = mtp_df.mean(axis=1)
```

```python
do_df = do_df[['시도명', '시군구명', 'annual_PM10']]
mtp_df = mtp_df[['시도명','annual_PM10']] 
```
