```python
df["month"] = df["측정일시"]
df["day"] = df["측정일시"]
df['year'] = df['측정일시']
```

```python
df['month'] = df['month'].apply(str)
df['day'] = df['day'].apply(str)
df['year'] = df['year'].apply(str)
```

```python
df.dtypes #슬라이싱을 위해 month, day, year을 문자열로 변환
```

```python
df['month'] = df['month'].str[4:6]
df['day'] = df['day'].str[6:8]
df['year'] = df['year'].str[:4]
```

```python
df
```

```python
df['시도명'] =df['주소'].str.split(" ",expand=True)[0]
df['시군구명'] =df['주소'].str.split(" ",expand=True)[1]
```

```python
df['시도명'] = df['시도명'].str[:2]
```

#### 측정소 코드가 같고 month가 같은 것끼리 묶어서 평균을 내야함

```python
df = df[['측정소코드', '시도명', '시군구명', 'year','month','day','PM10']]
```

```python
df
```

#### 측정소별 월별 평균 구하기

```python
monthly_df = pd.DataFrame(df.groupby(['측정소코드','month','시도명','시군구명'])['PM10'].mean()) ## 측정소 기준 월별 평균
```

```python
monthly_df = monthly_df.reset_index() #index --> column으로
```

```python
monthly_df = monthly_df.rename(columns = {'PM10': 'monthly_PM10'})
```

```python
monthly_df
```

#### 기존 df에 null 값 채우기

```python
df = pd.merge(df, monthly_df,on=['측정소코드','시도명','시군구명','month'])
```

```python
df
```

```python
df['PM10'] = df['PM10'].fillna(df['monthly_PM10'])
```

```python
df #NULL 값을 월별 평균으로 처리
```

#### 특별시와 광역시 따로 분류

```python
mtp_df = pd.DataFrame(monthly_df.groupby(['month','시도명'])['monthly_PM10'].mean())
do_df = pd.DataFrame(monthly_df.groupby(['month','시도명','시군구명'])['monthly_PM10'].mean())

mtp_df = mtp_df.reset_index() #index --> column으로
do_df = do_df.reset_index() #index --> column으로

mtp_df = mtp_df.loc[(mtp_df['시도명'] =='서울')|(mtp_df['시도명'] =='부산')|(mtp_df['시도명'] =='대구')|(mtp_df['시도명'] =='인천')|(mtp_df['시도명'] =='대전')|(mtp_df['시도명'] =='광주')|(mtp_df['시도명'] =='울산')|(mtp_df['시도명'] =='세종')]
do_df = do_df.loc[(do_df['시도명'] =='경기')|(do_df['시도명'] =='강원')|(do_df['시도명'] =='충남')|(do_df['시도명'] =='충북')|(do_df['시도명'] =='전남')|(do_df['시도명'] =='전북')|(do_df['시도명'] =='경남')|(do_df['시도명'] =='경북')|(do_df['시도명'] =='제주')]
```

```python
mtp_df
```

```python
do_df
```

```python
join = pd.merge(do_df, mtp_df, on=['month','시도명','monthly_PM10'], how= 'outer')
```
